# **Ex. 1 - First steps with scikit-learn**

Loading the Iris dataset from scikit-learn. Here, the third column represents the petal length, and the fourth column the petal width of the flower samples. The classes are already converted to integer labels where 0=Iris-Setosa, 1=Iris-Versicolor, 2=Iris-Virginica.

```python
from sklearn import datasets        # imports datasets from sci-kit learn package
import numpy as np                  # library that turns mathematical operations easier

iris = datasets.load_iris()         # the dataset we will work on
print("The dataset iris is of the type", type(iris))
print("These are its fields:", dir(iris))
print("Let's check it out...")
print(iris.DESCR)
```

```python
print(iris.data)
```

```python
X = iris.data[:, [2, 3]]    # we use slice to get all lines with only columns 2 and 3
                            # Column 2 = Petal Lenght
                            # Column 3 = Petal Width
y = iris.target             # Our target (the parameter we want to predict)        

print('Class labels:', np.unique(y))
# Y keeps the answers (classes). We want to know what classes we can have
# Using petal length and petal width, we will train a model to predict what class a flower is between 3 possibilities
```

Splitting data into 70% training and 30% test data:

```python
from sklearn.model_selection import train_test_split
# Now, we're going to train our model. First we have to prepare the data.
# We are going to use from the whole dataset, 70% of it for training and 30% for evaluating the result.

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1, stratify=y)
```

```python
# The np.bincount counts the total numbers each index appears inside a vector.
# For example:
my_vector = [1, 1, 2, 0, 4]
# we have the numbers 0, 1, 2 and 4 (we could know it using the unique method)
print(np.unique(my_vector))

# If we think the numbers as indexes of another vector, we can count the occurences of each one and store it in its respective index
np.bincount(my_vector)
```

```python
# We can use np.bincount to check if the number os ocurrences of a class is equal the sum of classes in y_train and y_test
print('Labels counts in y:', np.bincount(y))
print('Labels counts in y_train:', np.bincount(y_train))
print('Labels counts in y_test:', np.bincount(y_test))
```

Standardizing the features:

```python
from sklearn.preprocessing import StandardScaler
# Standardize features by removing the mean and scaling to unit variance.
sc = StandardScaler()

sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt


def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')

    # highlight test samples
    if test_idx:
        # plot all samples
        X_test, y_test = X[test_idx, :], y[test_idx]

        plt.scatter(X_test[:, 0],
                    X_test[:, 1],
                    c='',
                    edgecolor='black',
                    alpha=1.0,
                    linewidth=1,
                    marker='o',
                    s=100, 
                    label='test set')

X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))
```

# **Ex. 2 - Modeling class probabilities via logistic regression**

### *Part 1 - Train a logistic regression model* (check https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html) with scikit-learn: use C=100.0 and random_state=1. Plot the decision regions (use plot_decision_regions from mlxtend). Evaluate the results on the test set. 

```python
from sklearn.linear_model import LogisticRegression

# Initializes the model.
logReg = LogisticRegression (
    C = 100,
    random_state = 1)

# Training time.
logReg.fit(X_train, y_train)

# Plotting time. We already have the model ready to use.
plot_decision_regions(X, y, logReg, test_idx=None, resolution=0.02)

# Use score method to get accuracy of model,
score = logReg.score(X_test, y_test)
print(f'Accuracy of the model = {100*score:.2f}%')
```

### *Part 2 - Tackling overfitting via regularization*: evaluate different values for C (up to 100) and observe the accuracy difference between training and test sets. Plot the results.

```python
def accuracy(C_):
    logReg = LogisticRegression(C = C_, random_state = 1)
    logReg.fit(X_train, y_train)
    score = logReg.score(X_train, y_train)
    print(f'Accuracy of the model = {100*score:.2f}%')
    score = logReg.score(X_test, y_test)
    print(f'Accuracy of the model = {100*score:.2f}%')
```

# **Ex. 3 - Maximum margin classification with support vector machines**

### *Part 1 - Dealing with the nonlinearly separable case using slack variables* (check linear kernel at https://scikit-learn.org/stable/modules/svm.html): use C=1.0 and random_state=1. Plot the decision regions (use plot_decision_regions from mlxtend and included above). Evaluate the results on the test set. 

```python
from sklearn import svm
clf = svm.SVC(C=1.0, random_state=1)
clf.fit(X_train, y_train)
clf.predict([[2., 2.]])

plot_decision_regions(X_train, y_train, clf, test_idx=None, resolution=0.02)

# Use score method to get accuracy of model,
score = clf.score(X_test, y_test)
print(f'Accuracy of the model = {100*score:.2f}%')
```
