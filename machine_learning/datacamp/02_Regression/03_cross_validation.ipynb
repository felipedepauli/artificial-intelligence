```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# First of all, we have to load the dataframe to begin the work.
df = pd.read_csv("../datasets/gm_2008_region.csv")
sns.color_palette("Spectral", as_cmap=False)
sns.heatmap(df.corr(), cmap="Spectral")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

```python
reg = LinearRegression()
```

```python
X = df.drop('Region', axis=1).drop('life', axis=1).values
y = df['life'].values
```

```python
cv_scores = cross_val_score(reg, X, y, cv=5)
print("Scores \t\t:", cv_scores)
print(f"Average Score \t: {(np.mean(cv_scores)*100):.0f} %")
```
